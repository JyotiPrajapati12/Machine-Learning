#Handeling missing values in a **file**



```python
import pandas as pd
df=pd.read_csv("/weather_data.csv",parse_dates=["Day"])
df
```

```python
new_df=df.fillna(0)       #replace all the NaN value with 0
new_df
```

```python

new_df1=df.fillna({                    # in the form of a dictionary
    'temp':0,
    'windspeed':0,
    'event':'no event'
})
new_df1
```

```python
new_df2=df.fillna(method='ffill')            #Replace NaN values with the value from the previous row
#new_df2=df.fillna(method='ffill',limit=1)    Specifies the maximum number of NULL values to fill
new_df2
```

```python
new_df3=df.fillna(method='bfill')                            #Replace NaN values with the value from the next row
#new_df3=df.fillna(method='bfill',axis='columns')            Replace NaN values with the value from the next column
new_df3
```

```python
new_df4=df.dropna(how="all")      #drop the columns with all NaN values
#new_df4=df.dropna(how="any")      drop the columns with any NaN values
new_df4
```
